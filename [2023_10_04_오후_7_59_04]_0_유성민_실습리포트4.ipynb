```python
import numpy as np
import matplotlib.pyplot as plt
```


### (1) 선형회귀

- 문제: 시뮬레이션 데이터를 생성하여 선형회귀 분석 수행( w=3, b=4로 가정 )
- 실행결과: w와 b값이 각각 3.1984072584241137, 3.934023690751385로 근사한 최적값을 도출하였다. x가 1.5일 때 예측값은 8.731634578387556로 나타났다. 이터레이션이 진행되면서 오차값이 줄어드는 것을 볼 수 있다. 이터레이션에 따라 w와 b도 각각 3과 4에 수렴하는 것을 확인하였다.

```python
#시뮬레이션 데이터 생성
np.random.seed(128)
x = 2*np.random.rand(50,1)
y = 4+3*x+np.random.randn(50,1)
plt.scatter(x,y)

plt.show()
```

```python
#가중치 초기화
wi=np.random.uniform(low=-5.0, high=5.0)
bi=np.random.uniform(low=-5.0, high=5.0)

#이터레이션, 학습률 설정
num_iter=1000
learning_rate=0.01
```

```python
#가중치 업데이트
errors=[]
ws=[]
bs=[]

for iter in range(num_iter):
  y_hat = wi*x +bi
  wi = wi - learning_rate*((y_hat-y)*x).mean()
  bi = bi - learning_rate*(y_hat-y).mean()
  error=np.square(y_hat -y).mean()
  errors.append(error)
  ws.append(wi)
  bs.append(bi)
  if error < 0.01:
    break
```

```python
#최종 가중치 매개변수
wi, bi
```

```python
plt.scatter(x,y)
plt.plot([min(x),max(x)],np.array([min(x),max(x)])*wi+bi,"r-", label='predict')
plt.legend(loc="upper left", fontsize=10)

plt.show()
```

```python
#x=1.5일 때 예측값
print(wi*1.5 +bi)
```

```python
#이터레이션에 따른 오차값
plt.plot(errors)
plt.show()
```

```python
#이터레이션에 따른 w값
plt.plot(ws, label='weight')
plt.plot(bs, label='bias')
plt.axhline(3.0, color='lightgray', linestyle='--')
plt.axhline(4.0, color='lightgray', linestyle='--')
plt.axis([-100, 1100, 2.5, 4.5])
plt.legend(loc='upper left', fontsize=10)

plt.show()
```
