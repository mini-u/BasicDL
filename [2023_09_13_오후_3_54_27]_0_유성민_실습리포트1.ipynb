- 문제: 넘파이 벡터화 연산에 걸리는 시간과 리스트로 계산할 때 걸리는 시간 비교하기
- 실행결과: 도트 연산에 걸린 시간은 0.03169608116149902초, 리스트 0.48984265327453613초로 도트 연산이 0.45초 가량 빠름을 알 수 있다.

```python
import numpy as np
import time

#도트 연산
t1= time.time()
a = np.random.rand(1000000)
b = np.random.rand(1000000)
c = np.dot(a, b)
t2= time.time()

dot_time = t2-t1


#리스트
t3= time.time()
d = 0
for i in range(1000000):
 d += a[i]*b[i]
t4= time.time()

list_time = t4-t3


#걸리는 시간 비교
print(dot_time, list_time)
print(list_time-dot_time)
```

**실습3**
- 문제: sin, cos 함수 그리기
- 실행결과: 아래 그래프와 같이 나타났다. 사인 함수는 파란색으로, 코사인 함수는 빨간 점선으로 나타낼 수 있었다. y의 최저 및 최고값을 -1.2와 1.2로 조정하여 그래프에 여백을 주어 보기 편하게 하였다.


```python
import math
import matplotlib.pyplot as plt

x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b-', label='sin')
plt.plot(x, y2, 'r--', label='cos')
plt.title('sin & cos', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.2,1.2)
plt.legend()
plt.show()
```
